# LSA
The first method implemented is the LSA (Latent Semantic Analysis) or LSI (Latent Semantic Indexing). The function *LsiModel* is used to do so.


### Fit of the model

```python
from gensim.models import LsiModel
from gensim import corpora
from gensim.corpora import Dictionary
from gensim.models.coherencemodel import CoherenceModel
```

The function to fit the model requires as inputs the doc-term matrix and the dictionary of the words that appear in the corpus.

```python
docs_in_tokens = [doc.split() for doc in docs]
dictionary = corpora.Dictionary(docs_in_tokens)
doc_term_matrix = [dictionary.doc2bow(doc) for doc in docs_in_tokens]
```

Fit of the model:

```python
lsamodel = LsiModel(doc_term_matrix, num_topics=50, id2word = dictionary)
```

### Output
The function *print_topics* returns the first *t* linear combinations (topics) of the best r words associated to them, with the coefficient as well. *t* and *r* are parameters that the user can choose when calling the function.

```python
lsamodel.print_topics(num_topics=10, num_words=10)
```

### Evaluation
The *get_coherence* attribute of the *CoherenceModel* model from Gensim in used to compute the coherence metric.

```python
coherencemodel = CoherenceModel(model=lsamodel,
                                texts=docs_in_tokens,
                                dictionary=dictionary,
                                coherence='c_v')
coherencemodel.get_coherence()
```

# LDA model with sklearn
The first way we implemented LDA (Latent Dirichlet Allocation) is with Sklearn.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
```

A CountVectorizer is used. It is a text representation technique used in NLP to convert a collection of text documents into a numerical matrix that represents the frequency of words in those documents.

In practice, it breaks down sentences into tokens and counts the number of occurrences of each word within each document. This information is then organized into a matrix, where each row represents a document and each column represents a unique word found in the entire text corpus. The values within the matrix indicate how many times a particular word appears in a specific document.

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(docs)
```

```python
print(X.shape)
```

### Fit of the model
The function *LatentDirichletAllocation* is used. It needs the number of topics to be specified beforehand.

```python
lda = LatentDirichletAllocation(n_components=50, random_state=0)
lda.fit_transform(X)
```

### Visualization of the topics

```python
from __future__ import print_function
import pyLDAvis
import pyLDAvis.sklearn
import pyLDAvis.gensim
pyLDAvis.enable_notebook()
```

```python
vis = pyLDAvis.sklearn.prepare(lda, X, vectorizer)
```

```python
vis
```

### Evaluation
Since the *CoherenceModel* by Gensim cannot be used with this model, we used *metric_coherence_gensim* from the *tmtoolkit* library. It needs as inputs the number of words we want to consider for each topic, the components of the model, the vocabulary of the words in the corpus and the documents as a list of lists of the tokens.

```python
docs_in_tokens = [doc.split() for doc in docs]
```

```python
import tmtoolkit
coherence_topics = tmtoolkit.topicmod.evaluate.metric_coherence_gensim(measure='c_v', 
                        top_n=20, 
                        topic_word_distrib=lda.components_,
                        vocab=np.array([x for x in vectorizer.vocabulary_.keys()]), 
                        texts=docs_in_tokens)
```

```python
global_coherence = np.mean(coherence_topics)
global_coherence
```

# BERTopic
BERTopic is an open-source library for topic analysis based on the BERT algorithm.

### Preliminary steps

```python
import umap.umap_ as UMAP
from bertopic import BERTopic
```

We create here the Umap model which is later used to fit the BERTopic model:

```python
umap_model = UMAP.UMAP(n_neighbors=15, 
                  n_components=5, 
                  min_dist=0.0, 
                  metric='cosine', 
                  random_state=100)
```

### Fit of the model

Initiate BERTopic:

```python
topic_model = BERTopic(umap_model=umap_model, language="english", calculate_probabilities=True, nr_topics=50)
```

Run BERTopic model:

```python
topics, probabilities = topic_model.fit_transform(docs)
```

Get the list of topcs:

```python
topic_model.get_topic_info()
```

```python
topic_model.get_topic(7)
```

### Visualization of the topic top words

```python
topic_model.visualize_barchart(top_n_topics=12)
```

### Evaluation

```python
import numpy as np
from gensim.topic_coherence import direct_confirmation_measure
from gensim.models import CoherenceModel
```

```python
from gensim.corpora.dictionary import Dictionary

docs_in_tokens = [doc.split() for doc in docs]
dictionary = Dictionary(docs_in_tokens) 
```

```python
topic_words = [[words for words, _ in topic_model.get_topic(topic)] 
               for topic in range(len(set(topics))-1)]

# Evaluate
coherence_model = CoherenceModel(topics=topic_words, 
                                 texts=docs_in_tokens, 
                                 dictionary=dictionary, 
                                 coherence='c_v')
coherence = coherence_model.get_coherence()
coherence
```
